```python
output_filename = "raw_text.txt"
with open(output_filename, "w", encoding="utf-8") as f:
    f.write(raw_text)
```

```python
normalized_text = raw_text.replace('\n', ' ')# تبدیل line break به فاصله (تغییر فرمت داده)
```

```python
output_filename = "normalized_text.txt"
with open(output_filename, "w", encoding="utf-8") as f:
    f.write(normalized_text)
```

```python

# #جدا کردن جملات 
import re

# جایگزینی کلمات رایج با توکن موقت
abbreviations = ["Mr.", "Mrs.", "Ms.", "Dr.", "Prof.", "M.", "St."]
temp_token = "<DOT>"

for abbr in abbreviations:
    normalized_text = normalized_text.replace(abbr, abbr.replace(".", temp_token))

#  حذف _ های تاکید و line break ها
normalized_text = normalized_text.replace("_", " ").replace("\n", " ")

# تقسیم جملات
# - ابتدا با فاصله بعد از پایان جمله جدا می‌کنیم
# - سپس جملات را دوباره بررسی می‌کنیم تا کلمات درست شوند
sentence_splits = re.split(r'(?<=[.!?])\s+', normalized_text)
#  بازگرداندن نقطه در کلمات
sentence_splits = [s.replace(temp_token, ".") for s in sentence_splits]

# تمیز کردن فاصله‌ها و حذف جملات خالی
sentence_splits = [s.strip() for s in sentence_splits if s.strip()]
```

```python
#  ذخیره در فایل تکست
output_filename = "sentence_splits.txt"
with open(output_filename, "w", encoding="utf-8") as f:
    for sentence in sentence_splits:
        f.write(sentence + "\n")

print("تعداد جملات جدا شده:", len(sentence_splits))
```

```python
import re

normalized_sentences = []
punctuations = r"[.,!?;:\"'\(\)\[\]\{\}/\\|@#$%^&*]"

for sent in sentence_splits:
    sent = sent.lower()  # تبدیل همه حروف به کوچک
    
    temp = ""
    prev_space = False   # برای کنترل فاصله‌های پشت‌سرهم
    
    for ch in sent:
        # حذف عدد
        if re.fullmatch(r"[0-9]", ch):
            continue
        
        # حذف علائم نگارشی
        if re.fullmatch(punctuations, ch):
            continue
        
        # حذف کاراکترهای غیرحرفی (مثل ™ ”)
        if not re.fullmatch(r"[a-z ]", ch):
            continue

        # یکدست‌سازی فاصله‌ها
        if ch == " ":
            if not prev_space:
                temp += ch
            prev_space = True
        else:
            temp += ch
            prev_space = False
            
    temp = temp.strip()
    
    if len(temp) > 0:
        normalized_sentences.append(temp)
```

```python
with open("normalized_corpus.txt", "w", encoding="utf-8") as f:
    for s in normalized_sentences:
        f.write(s + "\n")
```

```python
final_sentences = []

for words in tokenized_sentences:
    new_sentence = ["<s>"] + words + ["</s>"]
    final_sentences.append(new_sentence)
```

```python
all_tokens = []

for sent in final_sentences:
    for token in sent:
        all_tokens.append(token)

#ساخت لیست کل توکن‌ها (برای unigram)
```

```python
vocabulary = []

for token in all_tokens:
    if token not in vocabulary:
        vocabulary.append(token)

#ساخت Vocabulary (کلمات یکتا)
```

```python
word_counts = {}

for token in all_tokens:
    if token in word_counts:
        word_counts[token] += 1
    else:
        word_counts[token] = 1

#شمارش فراوانی کلمات (پایه unigram)
```

```python
print(word_counts["<s>"])
print(word_counts["</s>"])
```

```python
total_tokens = 0

for token in all_tokens:
    total_tokens += 1

total_tokens
#تعداد کل توکن‌ها
```

```python
unigram_prob = {}

for word in word_counts:
    unigram_prob[word] = word_counts[word] / total_tokens

#محاسبه احتمال Unigram
```

```python
prob_sum = 0

for p in unigram_prob.values():
    prob_sum += p

prob_sum
#بررسی صحت مدل (جمع احتمال‌ها)
```

```python
import numpy as np

def generate_sentence_unigram(max_length=100):
    words = list(unigram_prob.keys())
    probs = list(unigram_prob.values())
    
    sentence = []
    
    while True:
        word = np.random.choice(words, p=probs)
        
        if word == "<s>":
            continue
        
        if word == "</s>":
            break
        
        sentence.append(word)
        
        if len(sentence) >= max_length:
            break
    
    return " ".join(sentence)
```

```python
import numpy as np

def generate_sentence_bigram2(bigram_prob, start_sentence=None, max_length=20):
    sentence = []

    if start_sentence:  # اگر جمله شروع داده شده
        words = start_sentence.split()
        sentence.extend(words)       # جمله شروع رو به جمله اضافه کن
        current_word = words[-1]     # از آخرین کلمه جمله شروع کن
    else:
        current_word = "<s>"

    while True:
        if current_word not in bigram_prob:
            break
        
        next_words = list(bigram_prob[current_word].keys())
        probs = list(bigram_prob[current_word].values())
        
        next_word = np.random.choice(next_words, p=probs)
        
        if next_word == "</s>":
            break
        
        sentence.append(next_word)
        current_word = next_word
        
        if len(sentence) >= max_length:
            break

    return " ".join(sentence)
```

```python
import numpy as np

def generate_sentence_trigram(trigram_prob, start_sentence, max_length=30):
    words = start_sentence.split(" ")
    
    sentence = words.copy()
    
    while True:
        if len(sentence) < 2:
            break
        
        context = (sentence[-2], sentence[-1])
        
        if context not in trigram_prob:
            break
        
        next_words = list(trigram_prob[context].keys())
        probs = list(trigram_prob[context].values())
        
        next_word = np.random.choice(next_words, p=probs)
        
        if next_word == "</s>":
            break
        
        sentence.append(next_word)
        
        if len(sentence) >= max_length:
            break
    
    return " ".join(sentence)
#تابع تولید جمله trigram
```

```python
print(generate_sentence_trigram(trigram_prob,start_sentence="only genius could have made charlotte"))
```

```python
import numpy as np

def generate_sentence_quadrigram(quadrigram_prob, start_sentence, max_length=40):
    words = start_sentence.split(" ")
    sentence = words.copy()
    
    while True:
        if len(sentence) < 3:
            break
        
        context = (sentence[-3], sentence[-2], sentence[-1])
        
        if context not in quadrigram_prob:
            break
        
        next_words = list(quadrigram_prob[context].keys())
        probs = list(quadrigram_prob[context].values())
        
        next_word = np.random.choice(next_words, p=probs)
        
        if next_word == "</s>":
            break
        
        sentence.append(next_word)
        
        if len(sentence) >= max_length:
            break
    
    return " ".join(sentence)
```
